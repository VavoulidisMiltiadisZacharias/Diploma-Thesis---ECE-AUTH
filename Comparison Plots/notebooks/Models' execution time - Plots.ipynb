### Extract names and values

```python
# Extract names and values
executionTimesNames = list(executionTimes.keys())
executionTimesValues = list(executionTimes.values())
```

### Plot execution times

```python
# Plot the training barchart
plt.figure(figsize=(12, 10))
bars = plt.bar(executionTimesNames, executionTimesValues, color='green', width=0.2)
plt.ylabel("Execution time (seconds)", fontsize = 12)
plt.title("Models' execution time", fontsize = 16)
plt.xticks(rotation=90, ha='right', fontsize = 12)  # Rotate labels for better readability

# Set y-axis limit (adjust max as needed to fit text)
plt.ylim(0, 10500)

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 150, f'{height:.4f}',
             ha='center', va='bottom', fontsize=12, rotation=90)


plt.tight_layout()
plt.show()
```

### Plot execution times (zoomed in)

```python
# Plot the zoomed in barchart
plt.figure(figsize=(12, 10))
bars = plt.bar(executionTimesNames, executionTimesValues, color='green', width=0.2)
plt.ylabel("Execution time (seconds)", fontsize = 12)
plt.title("Models' execution time (Zoomed in)", fontsize = 16)
plt.xticks(rotation=90, ha='right', fontsize = 12)  # Rotate labels for better readability

# Set y-axis limit (adjust max as needed to fit text)
plt.ylim(0, 400)

for bar, name in zip(bars, executionTimesNames):
    if name == "AdaBoost - No DR":
        continue  # Skip labeling this bar
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 5, f'{height:.2f}',
             ha='center', va='bottom', fontsize=10, rotation=90)

plt.show()
```

### Sort the execution times in ascending order

```python
# Sort by values (ascending)
sorted_executionTimes = dict(sorted(executionTimes.items(), key=lambda item: item[1]))
```
