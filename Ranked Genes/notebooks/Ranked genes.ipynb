### Compute how many times each gene appears in all the lists

```python
# Convert each list to a set to eliminate duplicates within a list
LowVarianceGenes = set(LowVarianceGenes)
LowMutualInfoDependencyGenes = set(LowMutualInfoDependencyGenes)
NNGGenes = set(NNGGenes)
MDGGenes = set(MDGGenes)

# Initialize counter for gene appearances across lists
gene_list_counts = Counter()

# Count how many lists each gene appears in
for gene_list in [LowVarianceGenes, LowMutualInfoDependencyGenes, NNGGenes, MDGGenes]:
    gene_list_counts.update(gene_list)

# Convert to DataFrame
df = pd.DataFrame(gene_list_counts.items(), columns=["Gene", "Frequency"])
df_sorted = df.sort_values(by="Frequency", ascending=False).reset_index(drop=True)

# Make sure full DataFrame prints
pd.set_option("display.max_rows", None)  # Show all rows
pd.set_option("display.max_columns", None)
pd.set_option("display.width", 1000)

# Print full DataFrame
print(df_sorted)
```
