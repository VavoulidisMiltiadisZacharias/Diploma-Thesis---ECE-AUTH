### Exclude last 2 columns (Targets)

### Store the dataset in an array

```python
X = df_excluded.values  # Convert DataFrame to NumPy array
M, N = X.shape          # M: number of patients, N: number of genes
```

### Set number of neighbors and d (number of genes to select)

```python
k = 3      # Number of nearest neighbors
d = 100     # Number of genes to select

# To store roughness score for each gene
result = np.zeros(N)  
```

### Step 1: Loop through each gene to compute roughness

```python
# Loop through each gene
for l in range(N):
    # Step 1.1: Get expression vector for gene l
    gene_signal = X[:, l]  # Shape (M,)

    # Step 1.2: Compute pairwise squared Euclidean distance between patients
    diff = gene_signal[:, np.newaxis] - gene_signal[np.newaxis, :]  # (M, M)
    distance_matrix = diff ** 2

    # Step 1.3: Construct k-NN graph adjacency matrix A_l for gene l
    A_l = np.zeros((M, M))
    for i in range(M):
        neighbors = np.argsort(distance_matrix[i])[:k + 1]  # include self
        neighbors = neighbors[neighbors != i][:k]           # exclude self
        A_l[i, neighbors] = 1

    # Step 1.4: Symmetrize adjacency matrix (make undirected)
    A_l = np.maximum(A_l, A_l.T)

    # Step 1.5: Compute Laplacian L_l = D - A
    D_l = np.diag(np.sum(A_l, axis=1))
    L_l = D_l - A_l

    # Step 1.6: Compute roughness score X_lᵀ L X_l
    result[l] = gene_signal.T @ L_l @ gene_signal
```

### Step 2: Sort results in descending order

```python
# Roughest first
sorted_indices = np.argsort(-result)

# Top d genes
selected_indices = sorted_indices[:d]
```

### Step 3: Build reduced gene expression matrix

```python
# Shape (M, d)
Y = X[:, selected_indices]

print(Y)
```

### Transform outliers

```python
# Function that tranforms outliers given a dataframe
def cap_outliers_iqr(df, threshold=1.5):
    df_capped = df.copy()
    for col in df.columns:
        if col in df_capped.columns and pd.api.types.is_numeric_dtype(df_capped[col]):
            Q1 = df_capped[col].quantile(0.25)
            Q3 = df_capped[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - threshold * IQR
            upper_bound = Q3 + threshold * IQR

            df_capped[col] = np.where(df_capped[col] < lower_bound, lower_bound, df_capped[col])
            df_capped[col] = np.where(df_capped[col] > upper_bound, upper_bound, df_capped[col])
        else:
            print(f"Warning: Column '{col}' not found or is not numeric. Skipping.")
    return df_capped
```

```python
# Transform outliers
df_capped = cap_outliers_iqr(df)
df_capped
```

### Check if "Transform Outliers" worked: Count how many elements are different between dataframes. If 0 nothing happened

```python
# Compare element-wise and count differences
num_differences = (df != df_capped).sum().sum()

# Print results
if num_differences > 0:
    print(f"The tranformation of outliers worked. The amount of outliers that has been transformed is: {num_differences}")
else:
    print("No values transformed during outliers transformation phase")
```
